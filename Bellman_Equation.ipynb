```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_s = 1 #reward for search
r_w = 0 #reward for wait
alpha = 0.5 #probability of staying in high battery level after search
beta = 0.5 # probability of successful search in low battery
gamma = 0.9 #discount factor which gives priority to immediate rewards or long term rewards. Close to 1 is for long term reward while 0 is immediate.
```

```python
results = []
v_h, v_l = 0,0
i=0
while i<100:

  v_h_1 = r_s + gamma*(alpha*v_h + ((1-alpha)*v_l))
  v_h_2 =  r_w + (gamma*v_h)


  v_l_1 = beta*r_s - 3*(1-beta)+gamma*(((1-beta)*v_h)+(beta*v_l))
  v_l_2 = r_w + gamma*v_l
  v_l_3 = gamma*v_h

  v_h = max(v_h_1,v_h_2)
  v_l = max(v_l_1, v_l_2, v_l_3)

  results.append([v_h,v_l])
  i+=1
```

```python
v_h_res, v_l_res = zip(*results)
plt.plot(v_h_res, label='v*(h)')
plt.plot(v_l_res, label='v*(l)')
plt.title('V*(h) and V*(l) values')
plt.xlabel('Iterations')
plt.ylabel('Values')
plt.legend()
plt.show()
```
